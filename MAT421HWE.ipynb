Continuity and Differentiation:

Continuity can generally be thought of as a function being connected over a given area. Given a function f(x), from x coordinates a to b, f(x) will be continuous if it is continued, or, in other words, connected from a to b on a graph.

Differentiability has to do moreso with integration, and being able to ascertain whether or not a point on a graph is able to be integrated at all. Following in with the visual of a graph, differentiability is one step above continuity; not only should f(x) be continous over an area, it should also not have any corners. Corners/points on a graph imply that f(x) is changing instantaneously, which is invalid for determining differentability.

Essentially, differentiability is determined by being able to predict where a graph is expected to go, based off of the point's differential equation. An instantaneous change of value is not able to be predicted by a (non-piecewise) differential equation.

This can be shown in code:

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-4,4,50)
yone=2*x
ytwo=1/x

plt.plot(x, yone,'-y')
plt.plot(x, ytwo,'-g')


```

The straight line is both continous and differentiable; it is defined along this area, and is smooth along it, with no corners or sharp points formed. This implies that, were one to take the differential equation of any point, one could reasonably determine where the "next" value for f(x) should be, with an incremental change in x.

While the green graph is continous, as it is defined across this area, it is not differentiable, as there is a sharp, instantaneous change near x = 0. This point, where the graph instantly goes from negative to positive, with only one x value, cannot be differentiated because one x value is being associated with many y values; the graph is connected over this area, but given any x value, one cannot predict future/current values of y (because one x value corresponds to many y values).

Taylor Theorem:
The taylor theorem is a polynomial method for approximating functions at/around a point. It creates a polynomial function that may not be accurate for the whole function that is desired to be replicated, but will, around the point desired, replicate the behavior of the desired function. Its general form is f(x) = f(a) + f'(a)(x-a) + f''(a)(x-a)^2/2! + f'''(a)(x-a)^3/3!..., where the accuracy can be increased by adding more terms (through differentiating f(x) again and again). It can be shown graphically as well:


```python
x = np.linspace(-np.pi, np.pi, 200)
y = np.zeros(len(x))

labels = ['First Order', 'Third Order']

plt.figure(figsize = (10,8))
for n, label in zip(range(2), labels):
    y = y + ((-1)**n * (x)**(2*n+1)) / np.math.factorial(2*n+1)
    plt.plot(x,y, label = label)

plt.plot(x, np.sin(x), 'k', label = 'Analytic')
plt.grid()
plt.title('Taylor Series Approximations of Various Orders')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

As can be seen, the sin function can be approximated with both a first order and third order taylor polynomial. The first order polynomial is a line, as it has only been differentiated to the first order, and the third resembles a cubic function. Outside of the point, these graphs clearly differ, but around the point, the polynomial approximates the desired sine function.

Optimization with gradient descent:
The gradient is essentially the derivative of a function. More specifically in vector calculus, it is the collection of partial derivatives that help form the gradient, but for these purposes, it is essentially synonymous with the derivative.

As the name of descent implies, it is also known as the direction of steepest descent, which is the direction in which the "point" will move fastest downward. In many cases, this is simply gravity, as gravity is the fastest way for things to go downward, relatively speaking.

So, gradient descent can be thought of as taking the differential of a function at a point, and using that differential to iterate downwards and downwards, until a function's minimum is found.

The reason this is used for optimization is that finding the minimum of a function is generally useful for finding floors and bases for mathematical processes, especially in the realm of vector calculus.

The general form for optimization with gradient descent is as follows:

f(n+1) = f(n) - a * f'(n)

f(n+1) represents the future value, f(n) represents the current value, a represents the learning rate, and f'(n) represents the gradient of the function at n, or, in other words, the differential of f evaluated at n.

At a glance, this equation very similar to a linear equation, and it indeed shares some similarities; this equation can broadly be thought of as taking a point, and making a very, infitesimally small jump downwards, and contiuing to do that until no more jumps can be made. This "jump" is amplified, or made faster/greater, by adjusting the learning rate, represented by a. By changing a, one can change how fast the approximation goes, by making jumps across bigger values. However, this also impacts the accuracy of the function, as greater learning rates make greater room for error.

Overall, this equation uses differentiation to find the fastest path downwards, and continually does that until a minimum is found.
