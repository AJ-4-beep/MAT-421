Numerical Differentiation Problem Statement:
A numerical grid is an evenly spaced set of points over the domain of a function; essentially, it is a collection of data points that are evenly spaced apart, most easily visualized as a set of x points for a given function f(x), for which the distance between one x data point and the next is a constant h. These are useful in some engineering and scientific analyses, and as such, the derivative of f(x) is a useful equation to estimate.

Thus, the problem statement is to estimate the derivative of f(x), in which the x inputs are aranged in a numerical grid. linspace is a simple way to create such a numerical grid:

```python
import numpy as np
x = np.linspace(1,2)
print(x)
```

Finite Difference Approximating Derivatives: The general form for the derivative/slope of f at a point a is as such: f'(a) = limx->a (  (f(x)- f(a))  / (x - a)). When x actually equals a, this equation takes on an indeterminate form; the limit is a way to estimate the behavior of f directly at a, by estimating the behavior of the function as f approaches a.

However, this means that there are 2 different ways to approach a; one can approach from the x data point one step before or after a; either moving forwards towards a, or backwards towards a in order to determine the expected behavior of the function. They are named to be the forward difference, for the line that connects a to the point in front of it, and the backward difference for the line that connects a to the point behind it.

Alternatively, the central difference is used for the line connected the points before and after a.

These lines are drawn to estimate the slope that actually exists at a, or to help estimate the derivative at a. This can be done more rigorously with Taylor series approximations for these difference equations.

Beacuse of how these difference equations are built with Taylor series, the central difference method will usually be the most accurate in determining the slope at a, as points directly around a, rather than simply on one side of a, are taken. This allows for greater accuracy in determining the actual derivative.

The general form for the central difference formula, with Taylor approximations used, is as such: f'(xj) = ( f(xj+1) - f(xj - 1) )  /  (2h)

This can be done in code as well, approximating the derivative of a function:

```python
import matplotlib.pyplot as plt

%matplotlib inline



h = 0.1

x = np.arange(0, 2*np.pi, h)
y = -1 * np.sin(x)

forward_diff = np.diff(y)/h

x_diff = x[:-1:]

exact_solution = -1 * np.cos(x_diff)

plt.figure(figsize = (12, 8))
plt.plot(x_diff, forward_diff, '--', \
label = 'Finite difference approximation')
plt.plot(x_diff, exact_solution, \
label = 'Exact solution')
plt.legend()
plt.show()

max_error = max(abs(exact_solution - forward_diff))
print(max_error)
```

The above code takes the differential of -sin, which is known to be -cos, and utilizes the difference approximation to estimate that -cos function. As can be seen, it is relatively accurate, but error still exists from the residual terms on the taylor series approximations. There is a finite difference that arises from the imperfection in the approximation, though it is relatively small.

Approximating of Higher Order Derivatives:

In the previous part, the first derivative's central difference approximation was established through subtracting higher order terms from each other, to have the first derivative (the slope) remain. However, rather than estimating the first derivative, the Taylor series' can be rearranged to estimate the second or third (or higher) derivatives. As an example, from the forward and backward difference equations, the second order derivative of f can be estimated as such: f''(xj) = ( f(xj+1) - 2f(xj) + f(xj -1)) / (h^2) (recall h being the step size/space between x points).

Approximating 2nd and higher order derivatives may seem without purpose, but these are useful in increasing the accuracy of the approximation, and understanding the behavior of f around the given point. Establishing concavity, convexity, and acceleration/deceleration, for example, can be very useful in certain engineering disciplines.
