Interpolation Problem Statement:

The Interpolation problem statement refers to the concept of applying an interpolation function to a given set of data. It's similar to regression, but is different in that regression necessitates an underlying model for the data; interpolation creates an "estimation function" for the data, such that, given a set of data points (xi, yi), with i being the set [1, ..., n], an estimation function y(x) can be created that satisfies y(xi) = yi for every data point.

In other words, interpolation is creating a function, that goes through the given data points, and connects them, so that we can predict what can happen at parts of the graph with no data points (such as places in between points where we have no given data).


Linear Interpolation:

Linear Interpolation is a method where lines are drawn between given data points, and the estimated points, that come from the function, must therefore lie on the line.

More specifically, say there is a x value called b that needs estimation (what is its y value?), and there is no given data point in the data that corresponds to b. Under Linear Interpolation, b's point will lie on the line connecting the nearest left and right points to b, which I'll call points a and c. Points a and c must actually exist in the given data points, as the line should estimate the behavior of the fucntion.

So, to estimate point b, we take its 2 nearest left and right points, which are indeed in the data set, and draw a line between these 2 points. We take b, its x value, and see what y value it matches to, to estimate what this point should be.

Below, see the linear interpolation for an x value of 1.2, given some existing data points.

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt



```

```python
x = [0, 1, 2]
y = [2, 4, 1]

f = interp1d(x,y)
y_i = f(1.2)
print(y_i)
```

This can be shown directly with a graph.

```python
plt.figure(figsize = (12,10))

plt.plot(x, y, '-ob')
plt.plot(1.2, y_i, 'ro')


plt.title('Linear Interpolation at x = 1.2')
plt.xlabel('x')
plt.ylabel('y')


plt.show()
```

Cubic Spline Interpolation:

Cubic Spline Interpolation is still interpolation, but just instead of lines being drawn between points, piecewise cubic functions are connected to create an estimation function.

More specifically, points are joined by the cubic polynomial yi(x) = aix^3 + bix^2 + cix + d, where ai, bi, ci, and di are constants, and must be determined. This cubic polynomial must be valid for xi <= x <= x(i+1) for i = [1,... ,n-1].

We can do this graphically as well:

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [0, 1, 2]
y = [2, 4, 1]

f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 2, 100)
y_new = f(x_new)
```

```python

plt.figure(figsize = (12,10))


plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')


plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')


plt.show()
```

As seen, part of a cubic function makes up this interpolation, connecting points.

So, interpolation helps estimates data based on given data points, and can be used, among other methods, with Linear Interpolation and Cubic Spline Interpolation.
